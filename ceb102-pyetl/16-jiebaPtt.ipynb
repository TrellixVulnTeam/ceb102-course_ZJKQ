```python
import jieba
import os
import glob
```

```python
path = 'res_gossiping'
os.listdir(path)
```

```python
glob.glob(path + '/*.txt')
```

```python
sb = "" # stringbuilder
for articlePath in glob.glob(path + '/*.txt'):
    with open(articlePath, 'r', encoding='utf-8') as f:
        sb += f.read().split('---split---')[0]
sb
```

```python
sbCut = jieba.cut(sb)
```

```python
wordList = [w for w in sbCut]
```

```python
# for i in sbCut:
#     print(i)
```

```python
wordList
```

```python
wordCountDict = dict()
for w in wordList:
    if w in wordCountDict:
        wordCountDict[w] += 1
    else:
        wordCountDict[w] = 1
wordCountDict
```

```python
wordCountList = [(k, v) for k, v in wordCountDict.items()]
wordCountList
```

```python
wordCountList.sort(key=lambda x: x[1], reverse=True)
```
